
Comprehensive Analysis of the Iris Dataset

```python
# Importing required libraries

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Reading the data
df = pd.read_csv("Iris.csv")
```

```python
# printing the data

print(df.shape)  # prints the number of rows and columns

# Displaying basic information about the dataset
print("\nSummary Statistics:")
print(df.describe())

# Displaying information about data types and missing values
print("\nData Types and Missing Values:")
print(df.info())
```

```python
# Determining the feature having null values and replacing them with zero
for col in df.columns:

    df[col] = df[col].replace(np.nan, 0)  # replacing the NAN with 0
    print(col, df[col].isnull().sum())   # to check the null values
```

```python

# Strip plot to visualize the relationship between species and petal length
plt.figure(figsize=(10, 6))
sns.stripplot(x="Species", y="PetalLengthCm", data=df, jitter=True, palette="husl")
plt.title("Strip Plot of Petal Length by Species")
plt.xlabel("Species")
plt.ylabel("Petal Length (cm)")
```

```python
# Pairplot for visualizing relationships between numerical features
sns.pairplot(df, hue="Species", markers=["o", "s", "D"], palette="husl")

# Adding title to the pairplot
plt.suptitle("Pairplot of Iris Dataset by Species", y=1.02)
```

```python
# Correlation heatmap for numerical features
numeric_columns = df.select_dtypes(include=[np.number]).columns
plt.figure(figsize=(10, 6))
sns.heatmap(df[numeric_columns].corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
```

```python
# Boxplot for each numerical feature by Species
plt.figure(figsize=(15, 8))
for i, col in enumerate(df.columns[:-1]):
    plt.subplot(2, 2, i % 4 + 1, label=f"Boxplot{i}")
    sns.boxplot(x="Species", y=col, data=df)
    plt.title(f"Boxplot of {col} by Species")

# Adding spacing between subplots
plt.tight_layout()

# Show the plots
plt.show()
```
